# Weapons used : Worldwide vs India

```python
weapons_worldwide=df['weaptype1_txt'].value_counts()[:5].index
values_weapons_worldwide=df['weaptype1_txt'].value_counts()[:5].values

weapons_india=df[df['country_txt']=='India']['weaptype1_txt'].value_counts()[:5].index
values_weapons_india=df[df['country_txt']=='India']['weaptype1_txt'].value_counts()[:5].values

plt.figure(figsize=(15,10))
plt.subplot(1, 2, 1)
plt.bar(weapons_worldwide,values_weapons_worldwide, color='red')
plt.title('Weapons mostly used Worldwide',fontsize=20)


plt.subplot(1, 2, 2)
plt.bar(weapons_india,values_weapons_india, color='#03a9f4')
plt.title('Weapons mostly used in India',fontsize=20)
plt.show()
```

### As we can see that in both cases the mostly used weapon comes in the form of Explosives

# Types of Target : Worldwide vs India

```python
targets_worldwide=df['targtype1_txt'].value_counts()[:5].index
values_targets_worldwide=df['targtype1_txt'].value_counts()[:5].values

targets_india=df[df['country_txt']=='India']['targtype1_txt'].value_counts()[:5].index
values_targets_india=df[df['country_txt']=='India']['targtype1_txt'].value_counts()[:5].values

plt.figure(figsize=(19,8))
plt.subplot(1, 2, 1)
plt.bar(targets_worldwide,values_targets_worldwide, color='red')
plt.title('Type of Targets Worldwide',fontsize=20)


plt.subplot(1, 2, 2)
plt.bar(targets_india,values_targets_india, color='#03a9f4')
plt.title('Type of Targets in India',fontsize=20)
plt.show()
```

### So we can see that in both the cases Private Citizens and property were targeted the most.

# Top 3 involved Terrorist Groups : Worldwide vs India

```python
groups_worldwide=df['gname'].value_counts()[:3].index
values_groups_worldwide=df['gname'].value_counts()[:3].values

groups_india=df[df['country_txt']=='India']['gname'].value_counts()[:3].index
values_groups_india=df[df['country_txt']=='India']['gname'].value_counts()[:3].values

plt.figure(figsize=(19,8))
plt.subplot(1, 2, 1)
plt.bar(groups_worldwide,values_groups_worldwide, color='red')
plt.title('Involved Terrorist Groups Worldwide',fontsize=20)


plt.subplot(1, 2, 2)
plt.bar(groups_india,values_groups_india, color='#03a9f4')
plt.title('Involved Terrorist Groups in India',fontsize=20)
plt.show()
```

### In both the cases the group causing the most attacks could not be known but apart from them Taliban is seen to be causing most of the problems worldwide whereas in india it is caused mostly by CPI-Maoist.
